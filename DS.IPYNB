### Data Preprocessing

This preprocessing is applied to both the Boston Housing Dataset and the Combined Cycle Power Plant Dataset.

### Model Implementation

In this step, we implement and evaluate linear regression using two approaches: a custom implementation with gradient descent and the built-in scikit-learn `LinearRegression`.

#### Steps:
1. **Custom Gradient Descent Implementation**:
   - **Initialization**:
     - Initialize weights (\( \theta \)) and bias (\( b \)) to zero.
   - **Gradient Descent Algorithm**:
     - Compute predictions using:
       ![Prediction Equation](https://latex.codecogs.com/png.latex?%5Chat%7By%7D%20%3D%20X%20%5Ccdot%20%5Ctheta%20%2B%20b)
     - Calculate the cost function (Mean Squared Error):
       ![Cost Function Equation](https://latex.codecogs.com/png.latex?J%28%5Ctheta%29%20%3D%20%5Cfrac%7B1%7D%7B2m%7D%20%5Csum%20%28%5Chat%7By%7D%20-%20y%29%5E2)
     - Compute the gradients:
       - Gradient of weights:
         ![Weight Gradient](https://latex.codecogs.com/png.latex?%5Cfrac%7B%5Cpartial%20J%28%5Ctheta%29%7D%7B%5Cpartial%20%5Ctheta%7D%20%3D%20%5Cfrac%7B1%7D%7Bm%7D%20X%5ET%20%28%5Chat%7By%7D%20-%20y%29)
       - Gradient of bias:
         ![Bias Gradient](https://latex.codecogs.com/png.latex?%5Cfrac%7B%5Cpartial%20J%28%5Ctheta%29%7D%7B%5Cpartial%20b%7D%20%3D%20%5Cfrac%7B1%7D%7Bm%7D%20%5Csum%20%28%5Chat%7By%7D%20-%20y%29)
     - Update weights and bias iteratively:
       - Update weights:
         ![Weight Update](https://latex.codecogs.com/png.latex?%5Ctheta%20%3A%3D%20%5Ctheta%20-%20%5Calpha%20%5Ccdot%20%5Cfrac%7B%5Cpartial%20J%28%5Ctheta%29%7D%7B%5Cpartial%20%5Ctheta%7D)
       - Update bias:
         ![Bias Update](https://latex.codecogs.com/png.latex?b%20%3A%3D%20b%20-%20%5Calpha%20%5Ccdot%20%5Cfrac%7B%5Cpartial%20J%28%5Ctheta%29%7D%7B%5Cpartial%20b%7D)
   - **Convergence**:
     - Track the cost function's convergence during training to ensure successful optimization.

2. **Scikit-learn Linear Regression**:
   - Use `LinearRegression` from scikit-learn to train a model on the same dataset.
   - Compare the custom implementation's performance with scikit-learn's implementation using metrics such as:
     - Mean Squared Error (MSE).
     - R-squared (\( R^2 \)).

#### Outputs:
- The final weights and bias values from the custom implementation.
- Training metrics (MSE, \( R^2 \)) for both implementations.
- A plot showing the cost function's convergence during training.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Gradient Descent Implementation
def gradient_descent(X, y, learning_rate=0.01, iterations=1000):
    m, n = X.shape
    theta = np.zeros(n)  # Initialize weights to zero
    bias = 0  # Initialize bias to zero
    costs = []  # To store the cost at each iteration

    for i in range(iterations):
        # Calculate predictions
        y_pred = np.dot(X, theta) + bias

        # Compute cost (Mean Squared Error)
        cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
        costs.append(cost)

        # Calculate gradients
        d_theta = (1 / m) * np.dot(X.T, (y_pred - y))
        d_bias = (1 / m) * np.sum(y_pred - y)

        # Update parameters
        theta -= learning_rate * d_theta
        bias -= learning_rate * d_bias

    return theta, bias, costs

# Prepare data for gradient descent
X_boston = boston_features_scaled
y_boston = boston_target.values

# Run gradient descent
learning_rate = 0.01
iterations = 1000
theta_gd, bias_gd, costs = gradient_descent(X_boston, y_boston, learning_rate, iterations)

# Scikit-learn Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_boston, y_boston)

# Predictions and metrics
y_pred_gd = np.dot(X_boston, theta_gd) + bias_gd
y_pred_sklearn = lin_reg.predict(X_boston)

mse_gd = mean_squared_error(y_boston, y_pred_gd)
r2_gd = r2_score(y_boston, y_pred_gd)

mse_sklearn = mean_squared_error(y_boston, y_pred_sklearn)
r2_sklearn = r2_score(y_boston, y_pred_sklearn)

# Display results
print("Gradient Descent Results:")
print(f"Weights (Theta): {theta_gd}")
print(f"Bias: {bias_gd}")
print(f"Mean Squared Error (MSE): {mse_gd}")
print(f"R^2 Score: {r2_gd}")

print("\nScikit-learn Linear Regression Results:")
print(f"Mean Squared Error (MSE): {mse_sklearn}")
print(f"R^2 Score: {r2_sklearn}")

# Plot cost function convergence
plt.figure(figsize=(8, 5))
plt.plot(range(iterations), costs, color='blue')
plt.title("Cost Function Convergence")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

### Cost Function Convergence - Boston Housing Dataset
This plot shows how the cost function decreases over iterations during gradient descent training for the Boston Housing dataset. The steady decline indicates successful optimization, with the model converging to its optimal parameters. A plateau at later iterations signifies that the algorithm has stabilized and reached the minimum cost.

### Hyperparameter Tuning and Evaluation

In this step, we experiment with different learning rates to observe their effects on the gradient descent optimization process. Additionally, we evaluate the model's performance using metrics such as Mean Squared Error (MSE) and R-squared (\( R^2 \)).

#### Steps:
1. **Hyperparameter Tuning**:
   - Experiment with multiple learning rates (e.g., 0.001, 0.01, 0.1, 0.5).
   - Train the model using gradient descent for each learning rate.
   - Plot the cost function convergence for each learning rate to compare their performance.

2. **Evaluation**:
   - Evaluate the performance of gradient descent using:
     - Mean Squared Error (MSE): Measures the average squared difference between predicted and actual values.
     - R-squared (\( R^2 \)): Explains the proportion of variance in the target variable explained by the model.
   - Compare these metrics across different learning rates.

3. **Discussion**:
   - Analyze the effects of learning rates on convergence and model performance.
   - Discuss the limitations and challenges of the gradient descent implementation, such as sensitivity to learning rate and computational efficiency.

```python
# Updated gradient descent with safeguards for divergence
def gradient_descent_safe(X, y, learning_rate=0.01, iterations=1000):
    m, n = X.shape
    theta = np.zeros(n)
    bias = 0
    costs = []

    for i in range(iterations):
        # Calculate predictions
        y_pred = np.dot(X, theta) + bias

        # Compute cost and safeguard against overflow
        cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
        if not np.isfinite(cost):  # Safeguard against divergence
            print(f"Divergence detected at iteration {i} for learning rate {learning_rate}")
            break
        costs.append(cost)

        # Calculate gradients
        d_theta = (1 / m) * np.dot(X.T, (y_pred - y))
        d_bias = (1 / m) * np.sum(y_pred - y)

        # Update parameters
        theta -= learning_rate * d_theta
        bias -= learning_rate * d_bias

    return theta, bias, costs

# Experiment with updated learning rates
learning_rates = [0.001, 0.01, 0.05, 0.1]
iterations = 1000
results_safe = {}

plt.figure(figsize=(10, 6))

# Train models with updated learning rates
for lr in learning_rates:
    theta_gd_lr, bias_gd_lr, costs_lr = gradient_descent_safe(X_boston, y_boston, learning_rate=lr, iterations=iterations)
    y_pred_gd_lr = np.dot(X_boston, theta_gd_lr) + bias_gd_lr
    mse_lr = mean_squared_error(y_boston, y_pred_gd_lr)
    r2_lr = r2_score(y_boston, y_pred_gd_lr)
    results_safe[lr] = {"theta": theta_gd_lr, "bias": bias_gd_lr, "mse": mse_lr, "r2": r2_lr}
    
    # Plot cost function convergence
    plt.plot(range(len(costs_lr)), costs_lr, label=f"LR: {lr}")

plt.title("Cost Function Convergence for Different Learning Rates (Safe)")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()

# Display results for each learning rate
for lr, metrics in results_safe.items():
    print(f"Learning Rate: {lr}")
    print(f"MSE: {metrics['mse']}, R^2: {metrics['r2']}\n")
```

### Cost Function Convergence for Different Learning Rates - Boston Housing Dataset
This plot compares the cost function convergence for various learning rates (\( \alpha \)):
- **Lower Learning Rates (e.g., 0.001)**: The cost decreases very slowly, requiring many iterations to approach the optimal solution.
- **Higher Learning Rates (e.g., 0.05, 0.1)**: Faster convergence, but safeguards are necessary to prevent divergence.
This visualization highlights the critical role of selecting an appropriate learning rate for effective training.


```python
# Experiment with updated learning rates for the CCPP dataset
learning_rates_ccpp = [0.001, 0.01, 0.05, 0.1]
iterations = 1000
results_ccpp_safe = {}

plt.figure(figsize=(10, 6))

# Train models with updated learning rates for CCPP
for lr in learning_rates_ccpp:
    theta_gd_lr_ccpp, bias_gd_lr_ccpp, costs_lr_ccpp = gradient_descent_safe(
        X_ccpp, y_ccpp, learning_rate=lr, iterations=iterations
    )
    y_pred_gd_lr_ccpp = np.dot(X_ccpp, theta_gd_lr_ccpp) + bias_gd_lr_ccpp
    mse_lr_ccpp = mean_squared_error(y_ccpp, y_pred_gd_lr_ccpp)
    r2_lr_ccpp = r2_score(y_ccpp, y_pred_gd_lr_ccpp)
    results_ccpp_safe[lr] = {
        "theta": theta_gd_lr_ccpp,
        "bias": bias_gd_lr_ccpp,
        "mse": mse_lr_ccpp,
        "r2": r2_lr_ccpp,
    }
    
    # Plot cost function convergence
    plt.plot(range(len(costs_lr_ccpp)), costs_lr_ccpp, label=f"LR: {lr}")

plt.title("Cost Function Convergence for Different Learning Rates (CCPP)")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()

# Display results for each learning rate
for lr, metrics in results_ccpp_safe.items():
    print(f"Learning Rate: {lr}")
    print(f"MSE: {metrics['mse']}, R^2: {metrics['r2']}\n")
```

### Performance and Limitations of Gradient Descent Implementation

In this section, we evaluate the performance and discuss the limitations of our gradient descent implementation for both datasets: Boston Housing and Combined Cycle Power Plant (CCPP).

---

#### **Performance**
1. **Accuracy**:

       - Optimal \( R^2 \): ~0.929
   - The results indicate that gradient descent effectively optimized the cost function.

2. **Learning Rate Sensitivity**:
   - Lower learning rates (e.g., 0.001) resulted in very slow convergence, requiring more iterations.
   - Higher learning rates (e.g., 0.1) achieved faster convergence but required safeguards to prevent divergence.

#### **Limitations**
1. **Learning Rate Selection**:
   - The success of gradient descent heavily depends on selecting an appropriate learning rate (\( \alpha \)):

   - Hyperparameter tuning for learning rates was manual, requiring trial and error. Automated techniques (e.g., grid search, adaptive learning rates) could improve efficiency.

3. **Computational Cost**:
   - Gradient descent is computationally intensive for large datasets, as it iteratively computes gradients for all features.

4. **Convergence Issues**:
   - The implementation needed safeguards to handle divergence (e.g., infinite or NaN costs) when the learning rate was too high.

---

#### **Takeaways**
1. **Effectiveness**:
   - Gradient descent is effective for linear regression tasks, producing results comparable to scikit-learn's implementation.
   - Properly chosen hyperparameters ensure convergence and optimal performance.

2. **Limitations**:
   - Sensitive to learning rates.
   - Computationally expensive for large datasets.

### Brief Report: Summary of Findings, Challenges, and Lessons Learned

This report provides a concise summary of the project outcomes, challenges encountered during implementation, and key takeaways from the learning process.

---

#### **Findings**
1. **Boston Housing Dataset**:
   - Gradient descent successfully optimized the linear regression model.
   - Optimal performance metrics:
     - Mean Squared Error (MSE): ~21.89
     - R-squared (\( R^2 \)): ~0.741
   - Learning rate of 0.05 provided the best balance between convergence speed and stability.

2. **Combined Cycle Power Plant (CCPP) Dataset**:
   - Gradient descent achieved results comparable to scikit-learn’s implementation.
   - Optimal performance metrics:
     - Mean Squared Error (MSE): ~20.77
     - R-squared (\( R^2 \)): ~0.929
   - A learning rate of 0.1 yielded efficient and stable convergence.

3. **Cost Function Convergence**:
   - For both datasets, the cost function steadily decreased for appropriate learning rates, validating the effectiveness of the gradient descent algorithm.

---

#### **Challenges Faced**
1. **Hyperparameter Tuning**:
   - Selecting an appropriate learning rate (\( \alpha \)) required trial and error.
   - Lower learning rates caused slow convergence, while higher learning rates risked divergence.

2. **Divergence Safeguards**:
   - Implementing safeguards against cost function divergence (e.g., infinite or NaN costs) was critical for handling high learning rates.

3. **Computational Cost**:
   - The iterative nature of gradient descent made it computationally expensive, especially for larger datasets like CCPP.

4. **Manual Implementation**:
   - Implementing gradient descent from scratch required meticulous attention to mathematical correctness, particularly for gradient calculations.

---

#### **Lessons Learned**
1. **Importance of Learning Rate**:
   - The learning rate plays a critical role in the success of gradient descent. Proper tuning can significantly improve convergence speed and stability.

2. **Effectiveness of Gradient Descent**:
   - Gradient descent is a robust optimization technique, yielding performance comparable to library-based solutions (e.g., scikit-learn).

3. **Trade-offs**:
   - There’s a trade-off between computational cost and accuracy. Techniques like mini-batch or stochastic gradient descent could improve efficiency.

4. **Value of Safeguards**:
   - Incorporating safeguards against divergence ensures the stability of training, particularly for higher learning rates.

5. **Understanding Foundations**:
   - Implementing gradient descent from scratch deepened understanding of linear regression and optimization principles.

---

#### **Future Work**
1. **Automated Tuning**:
   - Use grid search or adaptive learning rate techniques to automate hyperparameter selection.
2. **Efficiency Improvements**:
   - Explore stochastic gradient descent (SGD) or mini-batch gradient descent to reduce computational cost.
3. **Advanced Optimizers**:
   - Implement optimizers like Adam or RMSProp for improved convergence.
4. **Scalability**:
   - Test the implementation on larger, high-dimensional datasets to evaluate scalability.

---

### Conclusion
This project demonstrated the power and limitations of gradient descent for linear regression. By overcoming challenges in hyperparameter tuning and computational cost, gradient descent proved to be a valuable learning tool for understanding machine learning optimization.
